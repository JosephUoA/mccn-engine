```python
import matplotlib.pyplot as plt

from mccn.client import MCCN

ENDPOINT = "http://115.146.84.224:8082/"
collection = "ozbarley"
```

### Full dataset

```python
client = MCCN(ENDPOINT, collection, shape=100)
```

```python
ds = client.load()
```

```python
ds
```

```python
plt.imshow(ds.b01.values[1,:,:])
plt.show()
```

### Area1 

```python
from mccn.extent import GeoBoxBuilder


bbox =  [147.321145152223, -35.0498179211375, 147.321556920868, -35.049642182112]
area1_client1 = MCCN(ENDPOINT, collection, geobox = GeoBoxBuilder(4326).set_bbox(bbox).set_shape(100, 100).build())
```

```python
area1_ds = area1_client1.load()
```

```python
area1_ds
```

```python
plt.imshow(area1_ds.b01.values[1,:,:])
plt.show()
```
